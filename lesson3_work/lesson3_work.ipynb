```python
from BeiJingSubwaySpider import LineSpider,StationSpider
from Station import Station
import requests
```

# 获取数据

```python
base_url = 'https://baike.baidu.com'
node_url = '/item/%E5%8C%97%E4%BA%AC%E5%9C%B0%E9%93%81/408485'
line_spider = LineSpider('北京地铁', base_url+node_url)
line_links = line_spider.get_line_links()
```

```python
line_stations = {}
for line_name in line_links:
    station_spider = StationSpider(line_name, base_url+line_links[line_name])
    line_stations[line_name] = station_spider.get_station()
```

# 站点对应的路线与下标

```python
station_dict = {}
for line_name in line_stations:
    for station_name in line_stations[line_name]:
        if station_name in station_dict:
            station_dict[station_name].set_info(line_stations[line_name].index(station_name),
                         line_name)
        else:
            station = Station(station_name)
            station.set_info(line_stations[line_name].index(station_name),
                         line_name)
            station_dict[station_name] = station
```

```python
for station_name in station_dict:
    print(station_name,station_dict[station_name].info)
```

# 换乘路线与站点

```python
# 每条路线可以换乘的其他路线，以及该路线换乘对应的站点索引
def transfer(line):
    # transfer_line为可换乘的路线
    transfer_line = {}

    for line2 in line_stations:
        # transfer_index为换乘路线时站点的索引位置
        transfer_station = []

        for station in line_stations[line]:
            if line2 != line and station in line_stations[line2]:
                transfer_station.append(station)

        if transfer_station:
            transfer_line[line2] = set(transfer_station)

    return transfer_line
```

```python
# line_transfers记录每条路线换乘的方式
line_transfers = {}
for line in line_stations:
    line_transfers[line] = transfer(line)
```

# 计算同一条线上两站相隔站数

```python
# 判断是否为环线
def is_loop_line(line):
    return line_stations[line][0] == line_stations[line][-1]

# 走同一条路线经过的站数
def same_line_counts(line, station1, station2):
    counts = abs(line_stations[line].index(station1) -
                 line_stations[line].index(station2))
    if is_loop_line(line):
        return min(len(line_stations[line]) - 1 - counts, counts)
    else:
        return counts

# 已知起始点、换乘站、结束点、路线1、路线2，求起始点到结束点经过的站数
def two_line_counts(start_line, end_line, start_station, change_station, end_station):
    return same_line_counts(start_line, start_station, change_station) + same_line_counts(end_line, change_station, end_station)
```

# 换乘路线与经过站数

```python
def input_station_deal(station):
    return station if '站'==station[-1] else station+'站'
```

```python
def search_line(line_transfers, start_station, end_station):
    #     对'站'字的格式处理
    start_station = input_station_deal(start_station)
    end_station = input_station_deal(end_station)
    
    #   找到起始站、结束站所在路线与下标，可能有多条线与下标
    start_lines = station_dict[start_station].info
    end_lines = station_dict[end_station].info

    through_lines = []  # 用through_lines存储经过的(每条路线, 所有换乘站站名， 经过站数), 例：([lines], [station], station_count)
    result_lines = [] # 用result_lines存储有效路径、换乘站站名、经过站数
    
    for start_line in start_lines:
        if start_line in end_lines: # 如果结束站与起始站可以有同一条路线到达：
            station_count = same_line_counts(start_line, start_station, end_station)
            result_lines.append(([start_line], [start_station,end_station], station_count)) # 没有换乘
            
        else:
            through_lines.append(([start_line], [start_station], 0))
            while True:
                through_line, through_station, station_count = through_lines.pop(0) # 取出当前经过路径，当前站点，经过站数
                if len(through_line) >= 4:
                    break
                    
                current_line = through_line[-1] # 获取正在经历路线
                current_station = through_station[-1] # 获取上次换乘站名

                for next_line in line_transfers[current_line]:
                    for change_station in line_transfers[current_line][next_line]: # 返回字典，两条路线可能有多个对接点
                        if change_station in through_station: # 上次在该站换乘
                            continue
                        
                        if next_line in end_lines: # 如果下一条路线与终点所在路线一致
                            result_line = through_line + [next_line]
                            result_change_station = through_station + [change_station] + [end_station]
                            resulte_through_count = station_count + two_line_counts(current_line, next_line, current_station, change_station, end_station)
                            result_lines.append((result_line, result_change_station, resulte_through_count))
                            continue
                            
                        else:
                            new_line = through_line + [next_line]
                            new_through_station = through_station + [change_station]
                            new_through_count = station_count + same_line_counts(current_line, current_station, change_station)
                            through_lines.append((new_line, new_through_station, new_through_count))

    return result_lines
```

```python
search_line(line_transfers,'苹果园站','四惠东站')
```

# 输出数据处理

```python
import pandas as pd
def output_search_line(start_station, end_station):
    routes = search_line(line_transfers,start_station,end_station)
    routes_frame = pd.DataFrame(routes,columns=['line','station','station_count'])
    routes_frame['change_count'] = routes_frame['line'].apply(len)-1
    routes_frame_sort = routes_frame.sort_values(by=['station_count','change_count']).reset_index(drop=True)
#     按经过站数、换乘次数排序，选取前五条数据
    return routes_frame_sort[0:5]
```
